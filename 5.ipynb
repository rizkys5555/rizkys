Data Science Libraries
NumPy: Provides support for large, multi-dimensional arrays and matrices for numerical operations.
Pandas: Offers data structures like DataFrames for efficient data manipulation and analysis.
Matplotlib/Seaborn: Libraries for creating static and statistical data visualizations.
Scikit-learn: Provides simple and efficient tools for machine learning algorithms (classification, regression, clustering).
TensorFlow/PyTorch: Open-source machine learning frameworks widely used for deep learning tasks.
